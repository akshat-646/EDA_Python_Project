### **IMPORT LIBRARIES AND CSV FILE**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"//content/Diwali Sales Data.csv",encoding="unicode_escape")
```

```python
df.shape
```

```python
df.head(5)
```

### **DATA CLEANING PROCESS**

```python
df.info()
```

```python
df.drop(['unnamed1'],axis=1,inplace =True)
```

```python
df.drop(['Status'],axis=1,inplace =True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True)
```

```python
pd.isnull(df).sum()
```

```python
df["Amount"] =df["Amount"].astype("int")
```

```python
df["Age"] =df["Age"].astype("int")
```

```python
df["Orders"] =df["Orders"].astype("int")
```

```python
df["Amount"].dtypes
```

```python
df.columns
```

```python
df.rename(columns = {"User_ID" : "Customer_ID"})
```

```python
df.describe()
```

```python
df[["Age","Orders","Amount"]].describe()
```

###**GRAPHICAL REPRESENTATION OF DATA**

**ON THE BASIS OF GENDER**

```python
ax = sns.countplot(x="Gender",data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(10, 5))
sales = df.groupby(["Gender"],as_index = False)["Amount"].sum().sort_values(by="Amount",ascending =False)
ax = sns.barplot(x="Gender",y="Amount",data =sales )
for bars in ax.containers:
    ax.bar_label(bars)
```

**ON THE BASIS OF AGE GROUP**

```python
plt.figure(figsize=(10, 5))
age_group_order = ["0-17", "18-25", "26-35", "36-45", "46-50", "51-55", "55+"]
ax = sns.countplot(data=df, x="Age Group", order=age_group_order)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Distribution of Age Group ")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
ax = sns.countplot(data=df, x="Age Group", hue="Gender", order=age_group_order)
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Distribution of Age Group by Gender")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python
sales = df.groupby(["Age Group"],as_index = False,observed=False)["Amount"].sum().sort_values(by="Amount")
plt.figure(figsize=(20, 5))
ax = sns.barplot(x="Age Group",y="Amount",data =sales,order=age_group_order)
for bars in ax.containers:
    ax.bar_label(bars)
```

**QUARTILE**

```python
plt.figure(figsize=(12, 6))
q1_orders= df['Orders'].quantile(0.25)
q2_orders = df['Orders'].quantile(0.50)
q3_orders = df['Orders'].quantile(0.75)
plt.axvline(q1_orders, color='green', linestyle='-', linewidth=1.5, label=f'{q1_orders} 25th Percentile ({q1_orders:.2f})')
plt.axvline(q2_orders, color='blue', linestyle='-', linewidth=1.5, label=f'{q2_orders} 50th Percentile ({q2_orders:.2f})')
plt.axvline(q3_orders, color='red', linestyle='-', linewidth=1.5, label=f'{q3_orders} 75th Percentile ({q3_orders:.2f})')
plt.title('Orders Distribution with Quantile Annotations')
plt.xlabel('Orders')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
